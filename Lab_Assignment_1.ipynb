Q1 Develop a Menu driven program to demonstrate the following operations of Arrays
——MENU——-
1.CREATE
2. DISPLAY
3. INSERT
4. DELETE
5. LINEAR SEARCH
6. EXIT

```python
#include <stdio.h>
#include <stdlib.h>

#define MAX_SIZE 100

void create(int arr[], int *size) {
    printf("Enter the number of elements: ");
    scanf("%d", size);
    printf("Enter the elements: ");
    for (int i = 0; i < *size; i++) {
        scanf("%d", &arr[i]);
    }
}

void display(int arr[], int size) {
    if (size == 0) {
        printf("Array is empty.\n");
        return;
    }
    printf("Array elements: ");
    for (int i = 0; i < size; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
}

void insert(int arr[], int *size, int element, int position) {
    if (*size >= MAX_SIZE) {
        printf("Array is full. Cannot insert.\n");
        return;
    }
    if (position < 0 || position > *size) {
        printf("Invalid position. Cannot insert.\n");
        return;
    }
    for (int i = *size - 1; i >= position; i--) {
        arr[i + 1] = arr[i];
    }
    arr[position] = element;
    (*size)++;
}

void delete(int arr[], int *size, int position) {
    if (position < 0 || position >= *size) {
        printf("Invalid position. Cannot delete.\n");
        return;
    }
    for (int i = position; i < *size - 1; i++) {
        arr[i] = arr[i + 1];
    }
    (*size)--;
}

void linear_search(int arr[], int size, int element) {
    int found = 0;
    for (int i = 0; i < size; i++) {
        if (arr[i] == element) {
            printf("Element found at position %d.\n", i);
            found = 1;
            break;
        }
    }
    if (!found) {
        printf("Element not found in the array.\n");
    }
}

int main() {
    int arr[MAX_SIZE];
    int size = 0;
    int choice;
    int element, position;

    do {
        printf("\n---MENU---\n");
        printf("1. CREATE\n");
        printf("2. DISPLAY\n");
        printf("3. INSERT\n");
        printf("4. DELETE\n");
        printf("5. LINEAR SEARCH\n");
        printf("6. EXIT\n");
        printf("Enter your choice: ");
        scanf("%d", &choice);

        if (choice == 1) {
            create(arr, &size);
        } else if (choice == 2) {
            display(arr, size);
        } else if (choice == 3) {
            printf("Enter element to insert: ");
            scanf("%d", &element);
            printf("Enter position to insert: ");
            scanf("%d", &position);
            insert(arr, &size, element, position);
        } else if (choice == 4) {
            printf("Enter position to delete: ");
            scanf("%d", &position);
            delete(arr, &size, position);
        } else if (choice == 5) {
            printf("Enter element to search: ");
            scanf("%d", &element);
            linear_search(arr, size, element);
        } else if (choice == 6) {
            printf("Exiting program.\n");
        } else {
            printf("Invalid choice. Please try again.\n");
        }
    } while (choice != 6);

    return 0;
}
```

Q2 Design the logic to remove the duplicate elements from an Array and after the
deletion the array should contain the unique elements.

```python
#include <stdio.h>

int remove_duplicates(int arr[], int n) {
    if (n <= 1) {
        return n;
    }

    for (int i = 0; i < n; i++) {
        for (int j = i + 1; j < n; j++) {
            if (arr[i] == arr[j]) {
                for (int k = j; k < n - 1; k++) {
                    arr[k] = arr[k + 1];
                }
                n--;
                j--;
            }
        }
    }
    return n;
}

int main() {
    int arr[] = {1, 2, 2, 3, 4, 4, 4, 5, 5};
    int n = sizeof(arr) / sizeof(arr[0]);

    printf("Original array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");

    int new_size = remove_duplicates(arr, n);

    printf("Array after removing duplicates: ");
    for (int i = 0; i < new_size; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");

    return 0;
}
```

Q4 Implement the logic to
a. Reverse the elements of an array
b. Find the matrix multiplication
c. Find the Transpose of a Matrix

```python
#include <stdio.h>

#define ROWS1 2
#define COLS1 3
#define ROWS2 3
#define COLS2 2

void reverse_array(int arr[], int n) {
    int start = 0;
    int end = n - 1;
    while (start < end) {
        int temp = arr[start];
        arr[start] = arr[end];
        arr[end] = temp;
        start++;
        end--;
    }
}

void multiply_matrices(int A[][COLS1], int B[][COLS2], int C[][COLS2]) {
    for (int i = 0; i < ROWS1; i++) {
        for (int j = 0; j < COLS2; j++) {
            C[i][j] = 0;
            for (int k = 0; k < COLS1; k++) {
                C[i][j] += A[i][k] * B[k][j];
            }
        }
    }
}

void transpose_matrix(int A[][COLS1], int T[][ROWS1]) {
    for (int i = 0; i < ROWS1; i++) {
        for (int j = 0; j < COLS1; j++) {
            T[j][i] = A[i][j];
        }
    }
}

void display_matrix(int rows, int cols, int matrix[rows][cols]) {
    for (int i = 0; i < rows; i++) {
        for (int j = 0; j < cols; j++) {
            printf("%d ", matrix[i][j]);
        }
        printf("\n");
    }
}

int main() {
    int arr[] = {1, 2, 3, 4, 5};
    int n = sizeof(arr) / sizeof(arr[0]);
    printf("Original array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
    reverse_array(arr, n);
    printf("Reversed array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n\n");

    int A[ROWS1][COLS1] = {{1, 2, 3}, {4, 5, 6}};
    int B[ROWS2][COLS2] = {{7, 8}, {9, 10}, {11, 12}};
    int C[ROWS1][COLS2];
    multiply_matrices(A, B, C);
    printf("Result of matrix multiplication:\n");
    display_matrix(ROWS1, COLS2, C);
    printf("\n");

    int T[COLS1][ROWS1];
    transpose_matrix(A, T);
    printf("Transpose of the first matrix:\n");
    display_matrix(COLS1, ROWS1, T);
    printf("\n");

    return 0;
}
```

Q5 Write a program to find sum of every row and every column in a two-dimensional
array.  

```python
#include <stdio.h>

#define ROWS 3
#define COLS 3

int main() {
    int matrix[ROWS][COLS];

    printf("Enter elements of the %dx%d matrix:\n", ROWS, COLS);
    for (int i = 0; i < ROWS; i++) {
        for (int j = 0; j < COLS; j++) {
            scanf("%d", &matrix[i][j]);
        }
    }

    printf("\nSum of each row:\n");
    for (int i = 0; i < ROWS; i++) {
        int row_sum = 0;
        for (int j = 0; j < COLS; j++) {
            row_sum += matrix[i][j];
        }
        printf("Row %d sum: %d\n", i + 1, row_sum);
    }

    printf("\nSum of each column:\n");
    for (int j = 0; j < COLS; j++) {
        int col_sum = 0;
        for (int i = 0; i < ROWS; i++) {
            col_sum += matrix[i][j];
        }
        printf("Column %d sum: %d\n", j + 1, col_sum);
    }

    return 0;
}
```
